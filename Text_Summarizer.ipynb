```python
import torch
import gradio as gr


from transformers import pipeline

# path = "../models/huggingface/hub/models--sshleifer--distilbart-cnn-12-6/.no_exist/a4f8f3ea906ed274767e9906dbaede7531d660ff"

pipe = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6", )

```

```python
def Summary(text):
    output = pipe(text)
    return(output[0]["summary_text"])

gr.close_all()
# demo = gr.Interface(fn=Summary, inputs="text", outputs="text")
demo = gr.Interface(fn=Summary, inputs=gr.Textbox(label="Enter the text to be summarized", lines=10), outputs=gr.Textbox(label="Summarized Text", lines=6), title="Text Summarizer", description="THIS APPLICATION WILL BE USED TO SUMMARIZE YOUR TEXT")
demo.launch()
```
